# Bitwise Operators
- These operators act on individual bits (0 and 1) of the operands
- We can use bitwise operators directly on binary numbers or on a integers also
- When we use these operators on integers, these numbers are converted into bits and then bitwise operators act upon those bits
- Let x = 10 (0000 1010 in binary) and y = 4 (0000 0100 in binary)


```python
# Python program to understand bitwise operators 
a = 10            # 10 = 0000 1010 
b = 4            # 4 = 0000 0100 
c = 0

c = a & b;        # 0 = 0000 0000
print ("Line 1 - Value of c is ", c)

c = a | b;        # 14 = 0000 1110 
print ("Line 2 - Value of c is ", c)

c = a ^ b;        # 14 = 0000 1110 
print ("Line 3 - Value of c is ", c)

c = ~a;           # -11 = 1111 0101
print ("Line 4 - Value of c is ", c)

c = a << 2;       # 40 = 0010 1000
print ("Line 5 - Value of c is ", c)

c = a >> 2;       # 40 = 0000 0010
print ("Line 6 - Value of c is ", c)
```

# Identity Operators
- These operators compare the memory locations of two objects
- Hence it is possible to know whether the two objects are same or not
- The memory location of an object can be seen using the id() function. This function returns an integer number called the identity number that initialy represents the memory location of the object.


```python
a=25
b=25
print("The memory location of object a", id(a))
print("The memory location of object b", id(b))
```

- There are two identity operators
    1. <b>is
    2. <b>is not
- is operator is useful to compare whether two objects are same or not. It will internally compare the identity number of the objects. if the identity number of the objects are same, it will return True, otherwise it returns False.
- is not operator returns True, if the identity numbers of two objects being compared are not same. if they are same, then it returns false
- The 'is' and 'is not' operators do not compare the values of the objects. they compare the identity numbers or memory location of the objects

```python
# A Python program to understand the identity operators
a = 20
b = 20

if ( a is b ):
   print("a and b have same identity") 
else:
   print("a and b do not have same identity") 

b = 30
if ( a is b ):
   print("a and b have same identity")
else:
   print("a and b do not have same identity")

if ( a is not b ):
   print("a and b do not have same identity") 
else:
   print ("a and b have same identity") 
```
